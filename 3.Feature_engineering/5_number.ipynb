### Minimum, Maximum, Median, Q1, Q3, and IQR of a list of numbers

```python
import numpy as np
```

```python
lst_numbers = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimumn,Q1,median,Q3,maximum = np.quantile(lst_numbers,[0,0.25,0.50,0.75,1])
```

```python
print(minimumn,Q1,median,Q3,maximum)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5*IQR ## Value below the lower bound are outliers
upper_bound = Q3 + 1.5*IQR ## Value above the upper bound are outliers
```

```python
print(lower_bound, upper_bound)
```
