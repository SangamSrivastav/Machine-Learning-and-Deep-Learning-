```python
df1 = pd.DataFrame({
    'Key' : ['A','B','C'], 
    "Value" : [1,2,3]})

df2 = pd.DataFrame({
    'Key' : ['A','B','D'], 
    "Value" : [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='Key', how='outer')
```

```python
pd.merge(df1, df2, on='Key', how='inner')
```

```python
print(pd.merge(df1, df2, on='Key', how='left'))

print(pd.merge(df1, df2, on='Key', how='right'))
```
