```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip',data=tips)
plt.title('Scatter plot of total bill and tip')
```

```python
## Line plot
sns.lineplot(x='size', y='total_bill', data=tips) 
plt.title('Line plot of size and total bill')
```

```python
## Categorial plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of day and total bill')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.title('Bar plot of sex and total bill')
```

```python
## Box plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## Violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
 ## Histogram

sns.histplot(tips['total_bill'],kde=True)
plt.title('Histogram of total bill')

## kde = kernel density estimation
## kde is used to estimate the probability density function of a continuous random variable
```

```python
## KDE Plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pair plot

sns.pairplot(tips)
```

```python
## Heatmap

corr=tips[['total_bill','tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Heatmap of tips')

## annot is for showing the values in the heatmap
## cmap is for changing the color of the heatmap
## corr is for correlation
```
